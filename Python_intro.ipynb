## Циклы - повторение похожих действий

* Список - это хранилище для произвольного числа элементов. Список задается квадратными скобками: letters = ["a", "b", "c"]. Внутри элементы перечисляются через запятую.
* Цикл задается с помощью команды for, например: for number in [1, 2, 3]:
* Все, что выполняется внутри цикла, пишется с отступом в 4 пробела (или Tab)
* Цикл позволяет перебирать элементы последовательности (например, списка) и повторять для них одни и те же действия.
Пример:

```python
for number in [1, 2, 3]:
    print("Hello!")
    print(number)
```

## Подробнее про списки

В следующих заданиях Вам понадобятся операции со списками:
* Список задается квадратными скобками: letters = ["a", "b", "c"]. Внутри элементы перечисляются через запятую.
* Достать элемент из списка: letters[1] вернет "b". Нумерация идет с 0.
* Соединить все строки в списке через пробел: " ".join(letters) вернет "a b c".  Вместо пробела можно использовать любую строку.
* Добавить элемент в список: letters.append("d").

```python
letters = ["a", "b", "c"]
print(letters[1])
print(" ".join(letters))
```
